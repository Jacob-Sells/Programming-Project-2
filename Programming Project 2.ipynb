```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('james_bond_data_cleansed.csv')
```

```python
plt.scatter(data['imdb'],data['rotten_tomatoes'])
plt.title('Scatter Plot of Ratings')
plt.xlabel('Average IMDb Rating')
plt.ylabel('Average Rotten Tomatoes Rating')
plt.show()
print('There seems to be a correlation between both ratings, as both sites seem to agree on which movies are "better" than the others.')
```

```python
plt.scatter(data['imdb'],data['bond_kills'])
plt.title('Bond Kills vs IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Bond Kills')
plt.show()
print('There is no correlation between IMDb ratings and the number of kills in the movie.')
```

```python
x=(data['release_year'])
y1=(data['movie_budget'])
y2=(data['income_world'])
plt.plot(x,y1,label='Movie Budget')
plt.plot(x,y2,label='World Income')
plt.legend()
plt.show()
print('Over time, both the income and budget of the movies have increased. However, the income has increased more drastically.')
```

```python
ratio = (data['income_world']/data['movie_budget'])
plt.bar(data['release_year'], ratio, width = 0.75)
plt.xlabel('Release Year')
plt.ylabel('World Income-Movie Budget Ratio')
plt.title('World Income-Movie Budget Ratio Over Time')
plt.show()
print('Over time, the overall income of the movies has increased more than their budgets. It is likely that the producers began to make more money over time.')
```

```python
import statistics as stat
length_min = min(data['film_length'])
print(f"The shortest film length is {length_min}")
length_max = max(data['film_length'])
print(f"The longest film length is {length_max}")
length_mean = stat.mean(data['film_length'])
print(f"The average film length is {length_mean}")
length_med = stat.median(data['film_length'])
print(f"The median film length is {length_med}")

all_rating = (data['imdb'] + data['rotten_tomatoes'])
rating_min = min(all_rating)
print(f"The lowest film rating is {rating_min}")
rating_max = max(all_rating)
print(f"The highest film rating is {rating_max}")
rating_mean = stat.mean(all_rating)
print(f"The average film rating is {rating_mean}")
rating_med = stat.median(all_rating)
print(f"The median film rating is {rating_med}")
```
